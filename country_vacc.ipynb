```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
country_df = pd.read_csv('../Resources/country_vaccinations.csv')
country_df.head()
```

```python
country_df["total_vaccinations"]=country_df["total_vaccinations"].fillna(0)
country_df["people_vaccinated"]=country_df["people_vaccinated"].fillna(0)
country_df["people_fully_vaccinated"]=country_df["people_fully_vaccinated"].fillna(0)
country_df["daily_vaccinations_raw"]=country_df["daily_vaccinations_raw"].fillna(0)
country_df["daily_vaccinations"]=country_df["daily_vaccinations"].fillna(0)
country_df["total_vaccinations_per_hundred"]=country_df["total_vaccinations_per_hundred"].fillna(0)
country_df["people_vaccinated_per_hundred"]=country_df["people_vaccinated_per_hundred"].fillna(0)
country_df["people_fully_vaccinated_per_hundred"]=country_df["people_fully_vaccinated_per_hundred"].fillna(0)
country_df["daily_vaccinations_per_million"]=country_df["daily_vaccinations_per_million"].fillna(0)
```

```python
country_df.head()
```

```python
country_group=country_df.groupby("country").sum()
country_million=country_group["daily_vaccinations_per_million"]
country_sort=country_million.sort_values(axis=0, ascending=False)
country_top=country_sort.head(10)
country_top.head()
```

```python
x_axis = np.arange(len(country_top))
plt.bar(x_axis, country_top, color="b", align="center")
plt.show()
```
